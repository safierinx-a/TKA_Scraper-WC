```python
import bs4
import requests
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
url = "https://novelfull.com/the-kings-avatar/chapter-1-the-banished-battle-god.html"
```

```python
file1 = open(r"C:\Users\APOORV\Desktop\TKA.txt", "w+")
for j in range(1727):
    page = requests.get(url)
    soup = bs4.BeautifulSoup(page.content, 'lxml')
    raw = soup.find_all(name='p')
    for i in raw:
        try:
            file1.write(i.text)
            file1.write("\n")
        except:
            file1.write("\n")
    np = soup.find('a', attrs={'id':'next_chap'})
    url = 'https://novelfull.com' + np.get('href')
file1.close()
```

```python

stopwords = set(STOPWORDS) 
comment_words = ' '
file1 = open(r"C:\Users\APOORV\Desktop\TKA.txt", "r+")
x = file1.read()
# split the value 
tokens = x.split()     
# Converts each token into lowercase 
for j in range(len(tokens)): 
    tokens[j] = tokens[j].lower() 
for words in tokens: 
    comment_words = comment_words + words + ' '
```

```python
comment_words
```

```python

wordcloud = WordCloud(width = 1800, height = 1800, 
                background_color ='black', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
plt.savefig("TKA.svg")
```
